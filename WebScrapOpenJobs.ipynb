```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver
import json
from selenium import webdriver
from selenium.common.exceptions import TimeoutException

import time




# Configuramos el driver de Chrome
driver = webdriver.Chrome(executable_path=r'D:\crome driver\Crome\chromedriver-win64\chromedriver.exe')

# Indicamos la URL de la página que queremos scrape

#url = 'https://www.indeed.com/jobs?q=dentist&l=Dallas-Fort+Worth%2C+TX&sort=date&vjk=3989c381c9ebf3ba'
url= 'https://www.indeed.com/jobs?q=dentist&l=Dallas-Fort+Worth%2C+TX&sort=date&pp=gQAAAAABildkZP4AAAACD2pq3QADAAABAAA&vjk=e61fd325fc3404b5'
# Abrimos la página web
wait = driver.implicitly_wait(30)
driver.get(url)
job_data=[]
job_d=[]
company_d=[]
location_d=[]
date_d=[]


while True:
    try:
        #nombre de trabajo
        div = WebDriverWait(driver,100).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.resultContent')))
        #nombre de compania
        com = WebDriverWait(driver,100).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.companyName')))
        #locacion
        loc = WebDriverWait(driver,100).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.companyLocation')))
        #fecha
        date = WebDriverWait(driver,100).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.date')))
        
        for element in div:
            span = element.find_element(By.TAG_NAME,'span').text
            job_d.append(span)
           # print('job:',span)
        
        for a in com:
            company = a.text
            company_d.append(company)
         #   print('company:', company)
        for e in loc:
            location= e.text
            location_d.append(location)
          #  print('location',location)
        for d in date:       
            datee= d.text
            date_d.append(datee)
           # print('date: ',datee)
            
        #location of company
        
        # Encuentra y hace clic en el botón de "Siguiente página"
        siguiente_btn = WebDriverWait(driver,100).until(EC.element_to_be_clickable((By.CSS_SELECTOR, '[aria-label="Next Page"]')))
        siguiente_btn.click()                                                                
      
        # Espera a que la página se cargue completamente
        WebDriverWait(driver, 40).until(EC.presence_of_element_located((By.CSS_SELECTOR, '[aria-label="Next Page"]')))

        # Realiza tus acciones en la nueva página aquí, por ejemplo, obtén información de la página

    except NoSuchElementException:
        # Cuando no se puede encontrar el botón de "Siguiente página", significa que has llegado al final
        driver.quit()

        break

    except TimeoutException:
        # Cuando se lanza la excepción de tiempo de espera, significa que has llegado al final
        driver.quit()
        break


```

```python
import pandas as pd

data = {'Job': job_d, 'Company': company_d, 'Location': location_d, 'Date': date_d}

df = pd.DataFrame(data)
```
